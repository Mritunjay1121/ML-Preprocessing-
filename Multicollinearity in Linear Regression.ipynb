#### If correlation btw two features are > 90% ... then it means we are feeding same data .. this is an issue and is called as multicollinearuty

```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('Advertising.csv',index_col=0)
x=df.loc[:,['TV','radio','newspaper']]
y=df['sales']
df.head()
```

```python
x
```

```python
y
```

```python
# Using ordinary least square

x=sm.add_constant(x)
model=sm.OLS(y,x).fit()
```

```python
model.summary()
```

```python
x.iloc[:,1:].corr()
```

```python
df_salary=pd.read_csv('Salary_Data.csv')
df_salary
```

```python
X = df_salary[['YearsExperience', 'Age']]
Y = df_salary['Salary']
```

```python
## fit a OLS model with intercept on TV and Radio
X = sm.add_constant(X)
model_s= sm.OLS(Y, X).fit()
```

```python
model_s.summary()
```

```python
X.iloc[:,1:].corr()
```
