```python
ll=list(df3.columns)
for bb in ll:
    lst_10=df3[bb].value_counts().sort_values(ascending=False).head(10).index
   
    for a in lst_10:
        df3[bb+a]=np.where(df3[bb]==a,1,0)
df3
```

## Target Guided Ordinal Encoding


```python
import pandas as pd
df4=pd.read_csv('titanic-main.csv', usecols=['Cabin','Survived'])
df4
```

```python
df4['Cabin'].fillna('Missing',inplace=True)
df4
```

```python
print(df4['Cabin'].astype(str).str[0])
```

```python
df4['Cabin']=df4['Cabin'].astype(str).str[0]
```

```python
df4.head()
```

```python
df4.Cabin.unique()
```

```python
df4.groupby(['Cabin'])['Survived'].mean()
```

```python
df4.groupby(['Cabin'])['Survived'].mean().sort_values().index
```

```python
ordinal_labels=df4.groupby(['Cabin'])['Survived'].mean().sort_values(ascending=False).index
ordinal_labels
```

```python
for a in (enumerate(ordinal_labels,0)):
    print(a)
```

```python
ordinal_labels2={k:i for i,k in enumerate(ordinal_labels,0)}
ordinal_labels2
```

```python
df4['Cabin_ordinal_labels']=df4['Cabin'].map(ordinal_labels2)
df4.head(50)
```

## Mean Encoding - Nominal Encoding


```python
mean_ordinal=df4.groupby(['Cabin'])['Survived'].mean().to_dict()
```

```python
mean_ordinal
```

```python
df4['mean_ordinal_encode']=df4['Cabin'].map(mean_ordinal)
df4.head()
```

## Probability Encoding Steps

1.Probability of Survived based on Cabin--- Categorical Feature

2.Probability of Not Survived---1-pr(Survived)

3.pr(Survived)/pr(Not Survived)

4.Dictonary to map cabin with probability

5.replace with the categorical feature

```python
df5=pd.read_csv('titanic-main.csv',usecols=['Cabin','Survived'])
df5
```

```python
### Replacing
df5['Cabin'].fillna('Missing',inplace=True)
df5
```

```python
(df5['Cabin'].unique())
```

```python
df5['Cabin']=df5['Cabin'].astype(str).str[0]
df5.head()
```

```python
df5.Cabin.unique()
```

```python
prob_df=df5.groupby(['Cabin'])['Survived'].mean()
prob_df
```

```python
prob_df=pd.DataFrame(prob_df)
prob_df
```

```python
prob_df['Died']=1-prob_df['Survived'] # prob of not survived
```

```python
prob_df.head()
```

```python
prob_df['Probability_ratio']=prob_df['Survived']/prob_df['Died']
prob_df.head()
```

```python
probability_encoded=prob_df['Probability_ratio'].to_dict()
print(probability_encoded)
```

```python
df5['Cabin_encoded']=df5['Cabin'].map(probability_encoded)
df5.head(20)
```
